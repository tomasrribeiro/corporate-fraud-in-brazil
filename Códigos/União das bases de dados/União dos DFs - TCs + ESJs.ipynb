Agora que vimos os dois dataframes, vamos juntá-los com base na coluna "ID" e "Número do processo". No caso de não termos um match, vamos colocar em uma linha separada. Vale lembrar que isso é uma possibilidade. Digamos, por exemplo, que em um processo, todos os envolvidos tenham firmado TC. Nesse caso, a Sessão de Julgamento não acontece, e portanto, o Extrato dela não existe.

```python
from ast import literal_eval

df_tcs['identificador_lista_transformada'] = df_tcs['identificador_lista_transformada'].apply(literal_eval) #convert to list type
```

```python
exploded = df_tcs.explode('identificador_lista_transformada')
exploded
```

```python
print(df_esj['ID'].apply(type).unique())
```

```python
df_merged = df_esj.merge(exploded, left_on="ID", right_on="identificador_lista_transformada", how="outer")
```

Agora, temos um dataframe que combina os dois dados. Vale notar a pequena intersecção entre processos que tem compromitentes e ocorrência da Sessão de Julgamento. Isso era esperado, já que se um TC foi proposto, o processo é encerrado.

Vamos agora juntar as listas de acusados e compromitentes em uma só.

```python
n_linhas_nulas = df_merged[df_merged['Acusados'].isnull() & df_merged['compromitentes_lista_final'].isnull()].shape[0]

print(n_linhas_nulas)
```

```python
print(df_merged['Acusados'].apply(type).unique())
```

```python
print(df_merged['compromitentes_lista_final'].apply(type).unique())
```

```python
def combinar_listas(row):
    # Garantindo que os valores sejam sempre listas
    lista1 = row['Acusados'] if isinstance(row['Acusados'], list) else ([] if pd.isna(row['Acusados']) else [row['Acusados']])
    lista2 = row['compromitentes_lista_final'] if isinstance(row['compromitentes_lista_final'], list) else ([] if pd.isna(row['compromitentes_lista_final']) else [row['compromitentes_lista_final']])
    
    return lista1 + lista2

# Aplicando a função no dataframe
df_merged['acusados_compromitentes'] = df_merged.apply(combinar_listas, axis=1)
```

```python
df_merged
```

```python
total = df_merged['acusados_compromitentes'].apply(len).sum()

print(f"O total de acusados e compromitentes para os processos é {total}.")
```

Agora, temos o número final de acusados + compromitentes. São 8422 em 2094 processo, aproximadamente. Esses números, vale lembrar, não são finais, já que:

1. Na extração de dados dos ESJ, tivemos várias listas de acusados vazias.
2. Na extração de dados dos TCs, em casos pontuais tivemos cortes nos nomes graças às expressões regulares que utilizamos.

Apesar disso, parecem problemas residuais e podemos ter uma certa confiança nos números e dizer que devem aproximar bem o que vamos enfrentar.
